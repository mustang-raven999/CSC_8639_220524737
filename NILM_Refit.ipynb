```python
# Redefining the function to extract appliance activations from new_df

def extract_activations_from_new_df(df, appliance_name, power_threshold, min_on_duration, min_off_duration):
    """
    Extract activations for a given appliance from the new_df dataset.

    Parameters:
    - df: DataFrame
    - appliance_name: Name of the appliance column
    - power_threshold: Minimum power threshold to consider appliance as 'on'
    - min_on_duration: Minimum duration for which appliance should be 'on' to consider it an activation
    - min_off_duration: Minimum duration for which appliance should be 'off' between two activations

    Returns:
    - activations: List of dictionaries containing start and end Unix times for each activation
    """

    df_on = df[df[appliance_name] > power_threshold]

    activations = []

    start_time = -1
    end_time = -1
    previous_index = -1

    for index, row in df_on.iterrows():
        time = row['Unix']
        power = row[appliance_name]

        if start_time == -1:
            start_time = time
            end_time = time
        else:
            if time - end_time <= min_off_duration:
                end_time = time
            else:
                # save
                if end_time - start_time >= min_on_duration:
                    activations.append({
                        'start_time': start_time,
                        'end_time': end_time,
                        'duration': end_time - start_time,
                        'appliance_name': appliance_name
                    })
                start_time = time
                end_time = time

        previous_index = index

    if end_time - start_time >= min_on_duration:
        activations.append({
            'start_time': start_time,
            'end_time': end_time,
            'duration': end_time - start_time,
            'appliance_name': appliance_name
        })

    return activations

# Now, let's extract activations again using the updated mapping
all_activations_updated = {}

for appliance_name, params in params_appliance.items():
    column_name = appliance_mapping_updated.get(appliance_name, None)
    if column_name:
        activations = extract_activations_from_new_df(new_df,
                                                      column_name,
                                                      params['on_power_threshold'],
                                                      60,  # Using a fixed min_on_duration
                                                      60)  # Using a fixed min_off_duration
        all_activations_updated[appliance_name] = activations

# Displaying the first few activations for 'kettle' as an example
all_activations_updated['kettle'][:5]
```
